# Working with images
For analysing image data, we need to open them, apply filters to them, segment objects in the image and do measurements.

See also
* [Lecture notes on scikit image by Emmanuelle Gouillart](https://scipy-lectures.org/packages/scikit-image/index.html)

## Opening images
Most images read using the [imread](https://scikit-image.org/docs/dev/api/skimage.io.html#skimage.io.imread) function. In case your image doesn't open, you should consult the documentation of the given file format. Also check next section.

```python
from skimage.io import imread

image = imread('../../data/blobs.tif')
```

Images are just matrices of intensities. However, showing them as such is not convenient.

```python
image
```

## Image visualization
For visualizing images in notebooks, we can use the [imshow](https://scikit-image.org/docs/dev/api/skimage.io.html#skimage.io.imshow) function from the [scikit-image](https://scikit-image.org) library.

```python
from skimage.io import imshow

imshow(image)
```

Anothter example is the [imshow](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html) function from the [matplotlib](https://matplotlib.org/stable/index.html) library.

```python
import matplotlib.pyplot as plt

plt.imshow(image)
```

## Lookup tables (a.k.a. color maps)
We can also change the [look-up table](https://matplotlib.org/stable/tutorials/colors/colormaps.html), a.k.a. "color map" for the visualization. Available options can be found in the documentation of each function.

```python
imshow(image, cmap="hot")
```

```python
imshow(image, cmap="viridis")
```
