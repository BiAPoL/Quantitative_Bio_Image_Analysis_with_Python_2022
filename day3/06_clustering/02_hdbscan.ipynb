HDBSCAN allows to perform clustering of data points according to local neighborhood relations that are then weighted with distances between points. This allows to identify scarce points as noise and to assign cluster membership probabilities to each point.

```python
import numpy as np
import matplotlib.pyplot as plt
import hdbscan
from sklearn.datasets import make_blobs, make_moons
```

```python
n_samples=300
noisy_moons, y = make_moons(n_samples=n_samples, noise=0.1)
```

Plot the synthetic data first:

```python
fig, ax = plt.subplots()
ax.scatter(noisy_moons[:, 0], noisy_moons[:,1], c=y)
```

## Minimum cluster size

The minimum cluster size is a key parameter for hdbscan. It determines how small a cluster is allowed to be in order to be considered as a separate cluster. See the results of applying hdbscan with the default setting:

```python
clusterer = hdbscan.HDBSCAN()
clusterer.fit(noisy_moons)
```

Plot the prediction results. 

*Note*: We use the predicted labels (`clusterer.labels_`) to indicate each points's color and the cluster membership probabilities (`clusterer.probabilities_`) to indicate each point's transparency in the plot. The latter can be passed to the scatter plot through the `alpha` keyword argument:

```python
fig, ax = plt.subplots()
ax.scatter(noisy_moons[:, 0], noisy_moons[:,1], c=clusterer.labels_, alpha=clusterer.probabilities_)
```
